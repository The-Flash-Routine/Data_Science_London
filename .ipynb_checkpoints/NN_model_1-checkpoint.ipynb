```python
#for pandas 
import pandas as pd

#for numpy
import numpy as np

#for plotting purpose
import matplotlib.pyplot as plt
import seaborn as sns

#for keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.regularizers import l2

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
```

```python
def normalize_own(x):
    m=x.shape[0]
    mu=(1/m)*np.sum(x,axis=0)
    sig=np.std(x,axis=0,ddof=1)
    x=(x-mu)/sig
    return x,mu,sig
```

```python
def normalize_other(x,mu,sig):
    x=(x-mu)/sig
    return x
```

So there are no null values
So convert to numpy and then splitting training into pure train and pure validation

```python
xtrain=np.array(datatrainin)
ytrain=np.array(datatrainout)
xtrain,xval,ytrain,yval=train_test_split(xtrain,ytrain,test_size=0.2,random_state=91)
print(xtrain.shape,ytrain.shape)
print(xval.shape,yval.shape)
```

```python
#normalizing data and then adding bias of one
xtrain,mu,sig=normalize_own(xtrain)
xval=normalize_other(xval,mu,sig)
xtrain=np.hstack((np.ones(xtrain.shape[0]).reshape(-1,1),xtrain))
xval=np.hstack((np.ones(xval.shape[0]).reshape(-1,1),xval))
print(xtrain.shape)
print(xval.shape)
```

Now trying different types of NN's
Firstly without  Dropout

```python
model=Sequential()
model.add(Dense(80,input_dim=41,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(xtrain,ytrain,batch_size=10,epochs=100)
re=model.evaluate(xval,yval)
print(re)
```

Now lets vary number of units in 1st and 2nd hidden layer

```python
layer1=np.arange(10,100,10)
layer2=np.arange(2,30,5)
J_val=np.zeros((len(layer1),len(layer2)))
J_train=np.zeros(J_val.shape)
for i in range(0,len(layer1)):
    for j in range(0,len(layer2)):
      model=Sequential()
      model.add(Dense(layer1[i],input_dim=41,activation='relu'))
      model.add(Dense(layer2[j],activation='relu'))
      model.add(Dense(1,activation='sigmoid'))
      model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
      model.fit(xtrain,ytrain,batch_size=10,epochs=40,verbose=0)
      re=model.evaluate(xval,yval)
      J_val[i,j]=re[1]
      re=model.evaluate(xtrain,ytrain)
      J_train[i,j]=re[1]
```

```python
sns.set()
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.contour(layer2,layer1,J_val)
plt.xlabel("No of units in layer2")
plt.ylabel("No of units in layer1")
plt.subplot(1,2,2)
plt.contour(layer2,layer1,J_train)
plt.xlabel("No of units in layer2")
plt.ylabel("No of units in layer1")
```

```python
indmax_val=np.where(np.array(J_val)==np.max(J_val))
indmax_train=np.where(np.array(J_train)==np.max(J_train))
l1_val=indmax_val[0]
l2_val=indmax_val[1]
print(layer1[l1_val],layer2[l2_val])
```

Now searching near 60 and 12

```python
layer1=np.arange(56,64,1)
layer2=np.arange(8,16,1)
J_val=np.zeros((len(layer1),len(layer2)))
J_train=np.zeros(J_val.shape)
for i in range(0,len(layer1)):
    for j in range(0,len(layer2)):
      model=Sequential()
      model.add(Dense(layer1[i],input_dim=41,activation='relu'))
      model.add(Dense(layer2[j],activation='relu'))
      model.add(Dense(1,activation='sigmoid'))
      model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
      model.fit(xtrain,ytrain,batch_size=10,epochs=40,verbose=0)
      re=model.evaluate(xval,yval)
      J_val[i,j]=re[1]
      re=model.evaluate(xtrain,ytrain)
      J_train[i,j]=re[1]
```

```python
sns.set()
plt.figure(figsize=(15,5))
plt.contour(layer2,layer1,J_val)
plt.xlabel("No of units in layer2")
plt.ylabel("No of units in layer1")
```

```python
indmax_val=np.where(np.array(J_val)==np.max(J_val))
indmax_train=np.where(np.array(J_train)==np.max(J_train))
l1_val=indmax_val[0]
l2_val=indmax_val[1]
print(layer1[l1_val],layer2[l2_val])
```

So clearly we can take 61 units in layer 1  and 9 units in layer 2
Now lets take dropout into play

Clearly this is case of overfitting so now lets try with different values of Dropout

```python
dropset=np.arange(0.3,0.8,0.05)
J_val=[]
J_train=[]
for i in range(0,len(dropset)): 
  model=Sequential()
  model.add(Dense(61,input_dim=41,activation='relu'))
  model.add(Dense(9,activation='relu'))
  model.add(Dropout(dropset[i]))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  model.fit(xtrain,ytrain,batch_size=10,epochs=100,verbose=0)
  re=model.evaluate(xval,yval)
  J_val.append(re[1])
  re=model.evaluate(xtrain,ytrain)
  J_train.append(re[1])

plt.plot(dropset,J_val,c='green')
plt.plot(dropset,J_train,c='blue')
plt.xlabel("Drop Threshold")
plt.ylabel("Error")
plt.legend(["Validation_Error","Training Error"])
```

```python
So lets pick Dropout value to be 0.65
```

```python
model=Sequential()
model.add(Dense(61,input_dim=41,activation='relu'))
model.add(Dense(9,activation='relu'))
model.add(Dropout(0.65))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(xtrain,ytrain,batch_size=10,epochs=100,verbose=0)
```

Now we have to choose cutt-off/Threshold for sigmoid output unit

```python
epsilonset=np.arange(0.3,0.8,0.01)
J_val=[]
J_train=[]
for ep in epsilonset:
    pre=model.predict(xval)
    pre=(pre>ep).ravel()
    J_val.append(accuracy_score(yval,pre))
    pre=model.predict(xtrain)
    pre=(pre>ep).ravel()
    J_train.append(accuracy_score(ytrain,pre))
```

```python
plt.plot(epsilonset,J_train,c='blue')
plt.plot(epsilonset,J_val,c='green')
plt.xlabel("Threshold value")
plt.ylabel("Accuracy")
```

Lets try more variations of this networkas clearly the problem is of overfitting (from above graph)
Leta add l2 regularization to 2nd hidden layer

```python
l2set=np.arange(0.001,0,10)
J_val=[]
J_train=[]
for i in range(0,len(l2set)): 
  model=Sequential()
  model.add(Dense(61,input_dim=41,activation='relu'))
  model.add(Dropout(0.15))
  model.add(Dense(9,activation='relu',kernel_regularizer=l2(l2set[i])))
  model.add(Dropout(0.65))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  model.fit(xtrain,ytrain,batch_size=10,epochs=100,verbose=2)
  re=model.evaluate(xval,yval)
  J_val.append(re[1])
  re=model.evaluate(xtrain,ytrain)
  J_train.append(re[1])

plt.plot(l2set,J_val,c='green')
plt.plot(l2set,J_train,c='blue')
plt.xlabel("Drop Threshold")
plt.ylabel("Error")
plt.legend(["Validation_Error","Training Error"])
```

```python
Jv_nice=max(J_val)
Jt_nice=max(J_train)
print(Jv_nice)
l2_nice=l2set[np.argmax(J_val)]
print(l2_nice)
```
